## Linear Regression 

    var 1 = movie budget
    var 2 = movie revenue / pendapatan
    
    apakah semakin besar budget berpengaruh terhadap pendapatan?

## 1. Memanggil Library yang dibutuhkan

```python
import pandas
from pandas import DataFrame # untuk manipulasi data like membuat table, mengubah dimensi data dsb
import matplotlib.pyplot as plt # mebuat grafik plot
from sklearn.linear_model import LinearRegression # memanggil method2 yang sudah disediakan machine learning
```

```python
data = pandas.read_csv('cost_revenue_clean.csv')
data.describe()
```

## 2. Memanggil Dataset

```python
x = DataFrame(data, columns=['production_budget_usd'])
y = DataFrame(data, columns=['worldwide_gross_usd'])
```

## 3. Menampilkan grafik data asli

```python
plt.figure(figsize=(10,6))
plt.scatter(x, y, alpha = 0.3)
plt.title('Film cost vs global revenue')
plt.xlabel('Production Budget $')
plt.ylabel('worldwide gross $')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
```


## Modeling

```python
regression = LinearRegression()
regression.fit(x, y)
```

```python
#slope coef # tiap dollar yang diproduksi untuk pembuatan film menghasilkan 3 dollar revenue
regression.coef_
```

```python
#Intercept
regression.intercept_
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x, y, alpha = 0.3)
plt.plot(x, regression.predict(x), c='r', linewidth=4)

plt.title('Film cost vs global revenue')
plt.xlabel('Production Budget $')
plt.ylabel('worldwide gross $')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
```

```python
# predict revenue film dengan budget 50jt dollar #  x = intercept + coef * 50jt
# x = -7236192.7 + 3.1115 * 500000000 = 1548513807.3
regression.predict([[500000000]])
```
